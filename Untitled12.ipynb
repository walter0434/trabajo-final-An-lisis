```python
# Revisión de datos faltantes
print("Cantidad de registros con cero en el nivel de 'Age':", Datos["Age"].isin([0]).sum())
print("Cantidad de registros con cero en el nivel de 'SibSp':", Datos["SibSp"].isin([0]).sum())
print("Cantidad de registros con cero en el nivel de 'Survived':", Datos["Survived"].isin([0]).sum())
print("Cantidad de registros con cero en el nivel de 'Pclass':", Datos["Pclass"].isin([0]).sum())
```

```python
# remplezar todos los ceros con el valor medio (promedio) de la columna.
Datos["Age"] = Datos ["Age"]. replace(0,Datos["Age"] .mean())
Datos["SibSp"] = Datos ["SibSp"]. replace(0,Datos["SibSp"] .mean())
Datos["Survived"] = Datos ["Survived"]. replace(0,Datos["Survived"] .mean())
Datos["Pclass"] = Datos ["Pclass"]. replace(0,Datos["Pclass"] .mean())
```

```python
# Revision nuevamente de datos faltantes
print("Cantidad de registros con cero en el nivel de 'Age':", Datos["Age"].isin([0]).sum())
print("Cantidad de registros con cero en el nivel de 'SibSp':", Datos["SibSp"].isin([0]).sum())
print("Cantidad de registros con cero en el nivel de 'Survived':", Datos["Survived"].isin([0]).sum())
print("Cantidad de registros con cero en el nivel de 'Pclass':", Datos["Pclass"].isin([0]).sum())
```
